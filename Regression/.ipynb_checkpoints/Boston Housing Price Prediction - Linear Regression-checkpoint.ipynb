```python
X = ds.loc[:,'RM'].values.reshape(-1,1)
y = ds.loc[:,'MEDV'].values
```

```python
X[:2]
```

```python
y[:2]
```

# Split data into training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# 1. Lets try Simple Linear Regression using feature RM (Which has highest (positive) correlation)

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor_slm = LinearRegression()
```

```python
regressor_slm.fit(X_train,y_train)
```

```python
regressor_slm.intercept_
```

```python
regressor_slm.coef_
```

```python
regressor_slm.score(X_test,y_test)
```

```python
plt.figure(figsize=(10,10))
plt.plot(X_train,regressor_slm.predict(X_train))
plt.scatter(X_train,y_train)
plt.xlabel('average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes in $1000 s')
plt.grid()
```

```python
plt.figure(figsize=(10,10))
plt.plot(X_train,regressor_slm.predict(X_train))
plt.scatter(X_test,y_test)
plt.xlabel('average number of rooms per dwelling')
plt.ylabel('Median value of owner-occupied homes in $1000 s')
plt.grid()
```
