```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('housing.data', delim_whitespace=True, header=None)
```

```python
col_name = ['CRIM', 'ZN' , 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
df.columns = col_name
```

```python
df.head()
```

```python
#Refer linear regression for Exploratory data analysis and visualization
```

```python
#Lets try with one variable to understand RAnSAC, (since we can plot)
```

```python
X = df.loc[:,'RM'].values.reshape(-1,1)
```

```python
y = df.loc[:,'MEDV'].values
```

# Split data into training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

### http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RANSACRegressor.html

```python
from sklearn.linear_model import RANSACRegressor
```

```python
regressor_slRr = RANSACRegressor() 
```

```python
regressor_slRr.fit(X_train,y_train)
```

```python
regressor_slRr.score(X_test,y_test)
```

```python
regressor_slRr.score(X_train,y_train)
```

```python
inlier_mask = regressor_slRr.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,10));
plt.scatter(X_train[inlier_mask], y_train[inlier_mask], 
            c='blue', marker='o', label='Inliers')
plt.scatter(X_train[outlier_mask], y_train[outlier_mask],
            c='brown', marker='s', label='Outliers')
plt.scatter(X_test, y_test,
            c='green', marker='s', label='Test Data')
plt.plot(X_test,regressor_slRr.predict(X_test), color='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median value of owner-occupied homes in $1000's")
plt.legend(loc='upper left')
plt.show()
```

```python
regressor_slRr = RANSACRegressor(stop_score=.80) 
```

```python
regressor_slRr.fit(X_train,y_train)
```

```python
regressor_slRr.score(X_train,y_train)
```

```python
regressor_slRr.score(X_test,y_test)
```

```python
regressor_slRr.estimator_.coef_
```

```python
regressor_slRr.estimator_.intercept_
```

```python
inlier_mask = regressor_slRr.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

# Fit the model for all features

```python
X = df.iloc[:,0:13].values
```

```python
y = df.loc[:,'MEDV'].values
```

# Split data into training set and test set

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

# RANSAC Regression

```python
from sklearn.linear_model import RANSACRegressor
```

```python
regressor_RR = RANSACRegressor()
```

```python
regressor_RR.fit(X_train,y_train)
```

```python
regressor_RR.estimator_.intercept_
```

```python
regressor_RR.estimator_.coef_
```

```python
regressor_RR.score(X_train,y_train)
```

```python
regressor_RR.score(X_test,y_test)
```
